**1.What is a random variable in probability theory?**
Random variables are variable which can take different values based on random experimetn. Like tossing a coin or throwing a dice

**2: What are the types of random variables?**
Discrete Random variable: Which accepts only countable value
Continuous Random variable:It can take any numeric value from an interval
Categorical Random variable: Which takes qualitative data.

**3: Explain the difference between discrete and continuous distributions.**
A discrete distribution shows probablity distribution of discrete random variable outcome.
Continuous distribution shows the probablity distribution of continuous random variables outcome.

**4.What is a binomial distribution, and how is it used in probability?**
Binomial distribution is a discrete distribution which has only 2 possible outcome success or failure.

**5.What is the standard normal distribution, and why is it important?**
The standard normal distribution is a special case of the normal distribution (also called Gaussian distribution).It is important for calculating z score and central limit theorem.

**6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
When we take many random samples from any population (no matter the original distribution), the distribution of the sample means tends to become approximately normal (bell-shaped) as the sample size grows large.

**7.What is the significance of confidence intervals in statistical analysis?**
A confidence interval is a range of values that is likely to contain the true value of a population parameter (like mean, proportion, or difference between groups), with a certain level of confidence.

** 8: What is the concept of expected value in a probability distribution?**
The expected value of a random variable is the long-run average of its outcomes, weighted by their probabilities.


```python
'''9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.'''
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

rand_num = np.random.normal(loc = 50, scale = 5, size = 1000)
rand_num
df = pd.DataFrame(rand_num)
plt.hist(df, bins = 10)
plt.show()
```

```python
'''10: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval.'''

import numpy as np
import scipy.stats as stats
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

data = np.asarray(daily_sales)

mean_sales = np.mean(data)

n =len(data)

std_sales = np.std(data)

z_critical = stats.norm.ppf(0.975)

margin_of_error = z_critical*(std_sales/np.sqrt(n))

ci_lower =  mean_sales - margin_of_error
ci_upper = mean_sales + margin_of_error

print(f"Means sale is: {mean_sales}")
print(f"confidence interval is {ci_lower} to {ci_upper}")
```
