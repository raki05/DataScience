** 1: What is hypothesis testing in statistics?**
  
   Hypothesis testing in statistics is a method to make prediction/decision about population data based on sample data.

** 2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?**
  
   Null hypothsis is like default claim or nothing unusal or differnce is happening.
   Alternate hypothesis is challenging claim aginst null that something different or unusual has happened.

** 3:Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**
  
   The signigicance level is the level of threshold which we set that how much risk we are willing to take for type1 error. Signigicance level is reprsented by alpha and common value is 0.05.

** 4: What are Type I and Type II errors? Give examples of each.**
   
   Type 1 error is rejecting the null hypothisis when it's actually true.
   Type 2 error is failing to reject null hypothisis when it's actually false.

** 5: What is the difference between a Z-test and a T-test? Explain when to use
each.**
  
   Z-test is used when population variance is known and sample size is > 30.In z-test data follow the almost normal distribution when n value is large.

   T-test is used when population variance is unknown and sample size is < 30.Data comes from a normally distrbuted population.

```python
'''7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9] '''

import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

#let's assume null hypothesis is sample mean is 50

mu0 = 50

#alternate hypothesis mu_h != 50
# hence it's 2 tailed test

alpha = 0.05

data = np.array(sample_data)

n = len(data)

mean = np.mean(data)
std = np.std(data,ddof =1)

z_stat = (mean - mu0) / (std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))


if p_value < alpha:
  print(f"Reject Null hpyothesis i.e sample mean is significantly different from 50")
else:
  print(f"Failed to reject null hypothesis i.e no significant different from 50")






```

```python
''' 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.'''



import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# let's assume truemean ,std and size.
mu = 50
sigma = 5
n = 30

data = np.random.normal(mu, sigma, n)


sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
se = sample_std / np.sqrt(n)


confidence = 0.95
z_critical = stats.norm.ppf(0.975)
ci_lower = sample_mean - z_critical * se
ci_upper = sample_mean + z_critical * se

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Std Dev: {sample_std:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 4: Plot the data
plt.figure(figsize=(8,4))
plt.hist(data, bins=10)
plt.show()


```
