```python
#4.Write a python function that checks if a given nmber is prime or not from 1 to 200.
def fun(n):
  if n < 2 or n > 200:
    return False
  for i in range(2,n):
    if n % i == 0:
      return False
  return True

fun(199)
```

```python
#6.Write a generator function in python that yields power of 2 to a given exponent.

def func(n):
  for i in range(n+1):
    yield 2**i

value = func(5)
print(next(value))

for values in value:
  print(values)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
from docx import Document
def demo():
  doc = Document("DataType And Structure.docx")
  for para in doc.paragraphs:
    if para.text.strip():
       yield para.text

itr = demo()
print(next(itr))
print(next(itr))
```

```python
#8.Use a lambda function in python to sort a list of tuples based on the second element in each tuple:
data = [(1,3),(2,2),(4,1)]
sorted_data = sorted(data, key = lambda x:x[1])
print(sorted_data)
```

```python
#9.Write a python program that uses map() to convert a list of temperature form celciuss to farenhite.

celsius_temps = [0, 20, 37, 100]

fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))
print("Temperatures in Fahrenheit:", fahrenheit_temps)
```

```python
#10.Create a python program that uses filter() to remove all the vowels from a given string.

str = "pwskills"
vowels ="aeiou"

res = ''.join(filter(lambda ch : ch not in vowels,str))
print(res)
```
