```python
'''Question 6: Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]'''

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

x = np.array(list_x)
y = np.array(list_y)

mean_x = np.mean(x)
mean_y = np.mean(y)

cov_xy = np.mean((x-mean_x)*(y-mean_y))

corr_xy = np.corrcoef(x, y)[0, 1]

cov_xy, corr_xy

```

```python
'''Question 7: Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]'''

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

import matplotlib.pyplot as plt
import numpy as np
plt.boxplot(data, vert= False)
plt.xlabel("values")
plt.show()

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)


IQR = Q3 -Q1

Q1,Q3,IQR

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print( f"IQR is: {IQR}")
print( f"lower bound  is: {lower_bound}")
print( f"upper bound  is: {upper_bound}")
print( f"The outliers will lie beyond lower and upper bound hence outlier is: {outliers} ")


```

```python
'''Question 8: You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]'''

import numpy as np
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

cov_matrix = np.cov(advertising_spend,daily_sales,ddof=0)
covariance = cov_matrix[0,1]

correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print(f"covariance is: {covariance}")
print(f"correlation is: {correlation}")
print("Hence 2 variables are strongly corelated")
```

```python
''': Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]'''

# When analyzing survey score we would like to see central tendency and spread
# Histogram will show frequency distribution
#box plot help us found outliers and spread
# mean, median and mode scores will tell whether customers are generally satisfied or not
# std deviation will show where data is clustered and spread
#histogram will tell show the distribution

import matplotlib.pyplot as plt


import statistics

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

survey_mean = np.mean(survey_scores)
survey_median =  np.median(survey_scores)
survey_mode = statistics.mode(survey_scores)
survey_std = np.std(survey_scores)

survey_mean,survey_median,survey_mode,survey_std

plt.hist(survey_scores,bins = 10)
plt.show()
plt.boxplot(survey_scores,vert= False,)
plt.show()
```
