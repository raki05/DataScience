```python
#19 Write a code to find the union of 2 sets {1,2,3} and {3,4,5}.

s1 = {1,2,3}
s2 = {3,4,5}
union_set = s1.copy()
for ele in s2:
  union_set.add(ele)
print(union_set)
```

```python
#20. Write a code to find the intersection of 2 sets {1,2,3} and {2,3,4}

s1 = {1,2,3}
s2 = {2,3,4}
intersection_Set = set()
for ele in s1:
  if ele in s2:
    intersection_Set.add(ele)

print(intersection_Set)
```
