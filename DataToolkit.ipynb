**17. Explain the difference between apply() and map() in Pandas?**
  map() is primarily used with Series to map each element through a function, dictionary, or Series. It works element-wise and is mostly used for simple value-to-value transformations, like replacing or mapping values.
  apply() is more versatile and works on both Series and DataFrames. For Series, it applies a function element-wise (like map()). For DataFrames, it can apply a function along rows or columns (axis-wise), allowing more complex operations on rows or columns as a whole.

**20.What is the role of a pivot table in Pandas?**
  A pivot table in Pandas is used to summarize and aggregate data by reshaping a DataFrame. It allows you to group data by one or more keys (like categories or time periods), and then compute aggregated statistics (like sums, means, counts) for each group, organized in a tabular format.

**21.Why is NumPy’s array slicing faster than Python’s list slicing?**
  NumPy’s array slicing is faster than Python’s list slicing mainly because NumPy arrays are stored as contiguous blocks of memory with a fixed data type, allowing slicing to be handled as a view or a pointer to the original data without copying it. This means when you slice a NumPy array, it simply creates a new reference to the existing data, making it a very fast operation.

```python
'''4.A How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?'''

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 65000, 70000, 80000],
    'Experience': [1, 3, 5, 7, 9]
}

df = pd.DataFrame(data)


corr = df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
'''8.Create a line plot with multiple lines using Matplotlib?'''
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [5, 6, 7, 8]
y1 = [9, 10, 11, 12]

plt.plot(x, y, label='Line 1', color='blue', marker='o')
plt.plot(x, y1, label='Line 2', color='green', marker='s')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Multiple Lines')
plt.legend()
plt.grid(True)
plt.show()
```
